```python
import numpy as np
```

```python
import pandas as pd
```

```python
import glob
all_files = glob.glob("./*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
    
us_census = pd.concat(li, axis=0, ignore_index=True)
us_census
```

```python
us_census.dtypes
```

```python
us_census.head()
```

```python
us_census.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
us_census
```

```python
us_census.dtypes
```

```python
us_census[['male','female']] = us_census.GenderPop.str.split("_",expand=True)
```

```python
us_census.head()
```

```python
us_census.drop(columns=['GenderPop'], inplace=True)
```

```python
us_census.head()
```

```python
us_census['Income'] = us_census.Income.str.strip('$')
```

```python
us_census.head()
```

```python
us_census = us_census.replace('%','',regex=True)
```

```python
us_census
```

```python
us_census = us_census.replace('M','',regex=True)
us_census = us_census.replace('F','',regex=True)
```

```python
us_census.head()
```

```python
us_census.loc[:,'Hispanic':'Income'] = round(us_census.loc[:,'Hispanic':'Income'].apply(pd.to_numeric), 2)
```

```python
us_census.head()
```

```python
us_census.dtypes
```

```python
us_census.drop(columns=['location','price','product_type','product_description','quantity'], inplace=True)
```

```python
us_census.head()
```

```python
us_census
```

```python
us_census.dtypes
```

```python
np.where(us_census.applymap(lambda x: x == ''))
```

```python
us_census['male'] = us_census['male'].astype(float)
```

```python
us_census.dtypes
```

```python
us_census
```

```python
us_census.drop(columns=['female'], inplace=True)
```

```python
us_census
```

```python
us_census['female']= us_census['TotalPop']-us_census['male']
us_census['female']
```

```python
us_census
```

```python
us_census['male'] = us_census['male'].fillna(0).astype(int)
```

```python
us_census['male'] = us_census['male'].astype(int)
```

```python
us_census['female'] = us_census['female'].fillna(0).astype(int)
```

```python
us_census['female'] = us_census['female'].astype(int)
```

```python
us_census.dtypes
```
